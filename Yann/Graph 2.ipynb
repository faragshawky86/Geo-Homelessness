```python
Population_df = pd.read_csv("dataframes/State_Population.csv")
```

```python
Population_df.head()
```

```python
Yearlist = Population_df.columns[1:]
```

```python
for t in Population_df.columns[1:]:
    Population_df = Population_df.rename(columns={t: t + ' Population'})
    
# Rename state column
re_Population_df = Population_df.rename(columns={'STATE':'State'})
```

```python
re_Population_df.head()
```

```python
merged_homeless_df = pd.merge(re_Population_df, ann_home_df, on='State')
merged_homeless_df.head()
```

```python
for year in Yearlist:
    merged_homeless_df['homelessnessPercentage{}'.format(year)] = (merged_homeless_df[year] / merged_homeless_df[year + ' Population']) * 100
```

```python
merged_homeless_df.head()
```

```python
home_percentage_df = merged_homeless_df[['State', 'homelessnessPercentage2010', 'homelessnessPercentage2011', 'homelessnessPercentage2012',
                                        'homelessnessPercentage2013', 'homelessnessPercentage2014', 'homelessnessPercentage2015',
                                        'homelessnessPercentage2016', 'homelessnessPercentage2017']]
```

```python
home_percentage_df.head()
```

```python
home_percentage_df.to_csv("dataframes/Homeless_Percentage_by_Year.csv", header=True)
```
